```python
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return text

# Открываеv PDF-файл и извлекаем текст постранично
```

```python
def pars_pdf(text):
    text2 = text.split('\n')
    print(text2)
    if 'Рейтинг' in text2:
        num = text2.index('Рейтинг')
        raiting = text2[num-1]
        name = text2[num+1]
        activity = text2[num+2]
        date_publish = text2[num+3]
        print(num, raiting, name, activity, date_publish)
        
    else:
        raiting = 0
        name = ""
        activity = ""
        if 'Простой' in text2:
            num2 = text2.index('Простой')
            date_publish = text2[num2-1]
        else:
            date_publish = ""
        print(num, raiting, name, activity, date_publish)
    
    return  name, raiting, activity, date_publish

#  Функция принимает текст одной страницы и делит текст на строки с помощью '\n'.
# Если в списке строк есть слово 'Рейтинг', то извлекаем данные рядом с эти словом.
```

```python
df_corrected = pd.DataFrame({
    'CompanyName': NameCompany,
    'RatingValue': Raiting,
    'PublishInfo': DataPublish,
    'CompanyActivity': Activity,
    'ArticleText': TextArticle
})

def clean_publish_info(text):
    if "Автор оригинала:" in text:
        return None 
    if "Подписаться" in text or "Подпихаться" in text:
        return None
    return text

df_corrected['PublishDate'] = df_corrected['PublishInfo'].apply(clean_publish_info)
df_corrected['AuthorInfo'] = df_corrected['PublishInfo'].apply(
    lambda x: x if "Автор оригинала:" in str(x) else None
)

df_corrected.drop('PublishInfo', axis=1, inplace=True)

df_corrected['RatingValue'] = df_corrected['RatingValue'].astype(str).str.extract(r'(\d+\.\d+)')[0].astype(float)

def clean_article_text(row):
    text = row['ArticleText']
    company = row['CompanyName']
    rating = row['RatingValue']
    
    text = str(text).replace(f"{rating} Рейтинг {company}", "").strip()
    return text

df_corrected['ArticleText'] = df_corrected.apply(clean_article_text, axis=1)

df_corrected = df_corrected[['CompanyName', 'RatingValue', 'PublishDate', 'AuthorInfo', 
                            'CompanyActivity', 'ArticleText']]

df_corrected.to_csv("corrected_dataset.csv", index=False, encoding='utf-8-sig')

df_corrected.head()
```
